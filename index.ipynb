# Introduction to Project



Welcome to the project on Mr.HelpMateAI!



First and foremost, congratulations on completing the modules on semantic search, vector databases, and retrieval augmented generation! We hope this course offered a rich learning experience to you.


**What's next?**

Now that you are well versed with the fundamentals of building generative search systems, it's time to get your hands dirty and build one from scratch. 



In the search systems you have seen so far, there were essentially 3 layers: the emebdding layer, the search layer, and the generation layer. Each of these layers is open to experimentation in multiple avenues. Your goal in this session is to build a robust search system that can answer user queries effectively by experimenting with the various blocks in the system, such as chunking strategies, embedding models, re-rankers, generation prompts, etc.



In the next segment, you will understand the project in detail.
